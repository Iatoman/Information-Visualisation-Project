```python
import pandas as pd
import numpy as np
import csv
```

```python
# load dataset
df = pd.read_csv("Cleaned_responses_medianImputation.csv", index_col=False)
names = df.columns

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 120)
df.head()
# print(names)
```

```python
def med(x):
    length = len(x)
    x.sort()
    if (length % 2)== 1:
        z=length // 2
        y = x[z]
    else:
        y = (x[length//2]+x[length//2-1])/2
    if(y<0):y=-y
    return y

```

```python
def findList(time, list):
   indexnum = []
   for index in list:
        if(time in df.loc[index, 'time']):
            # print(df.loc[index, 'time'])
            indexnum.append(index)
   return indexnum       
```

```python
def findMed(time, feature, location):
    tmplist = df[df['location'] == location].index
    index = findList(time, tmplist)
    median = med(df.loc[index, feature].to_list())
    return median
```

```python
print(findMed('2020-04-08', 'power', 1))
print(findMed('2020-04-08', 'power', 2))
print(findMed('2020-04-08', 'power', 3))
```

```python
timeList = ['2020-04-06', '2020-04-07', '2020-04-08', '2020-04-09', '2020-04-10']
featureList = ['sewer_and_water', 'power', 'roads_and_bridges', 'medical', 'buildings', 'shake_intensity']

def getMedinSameday(time):
    finalList = []
    i=1
    while(i < 20):
        row = [time]
        # row = []
        for feature in featureList:
            row.append(findMed(time, feature, i))
        row.append(i)
        i = i + 1
        finalList.append(row)
    #print(finalList)
    return finalList
```

```python

result =[]
for time in timeList:
    result = result + getMedinSameday(time)
print(result)
```

```python
# Save the result to csv file:
name=['time', 'sewer_and_water', 'power', 'roads_and_bridges', 'medical', 'buildings', 'shake_intensity','location']
name2= range(1,96)
test=pd.DataFrame(columns=name, data=result)
print(test)
# 转换csv
test.to_csv('result.csv', encoding= 'gbk')
```
